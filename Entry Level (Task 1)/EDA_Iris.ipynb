# Iris Dataset Exploratory Data Analysis (EDA)

### 1. Import Relevant Libraries/Packages

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

### 2. Load Dataset and Create Dataframe

```python
df = pd.read_csv('Iris.csv')
df.head()
```

### 4. Exploratory Data Analysis (EDA) or Data Visualization

```python
sns.pairplot(data=df, hue='Species')
plt.show()
```

```python
# feature distribution
plt.figure(figsize=(15, 10))
plt.title('Feature Distribution using Violin Plot')
plt.subplot(2, 2, 1)
sns.violinplot(x='Species', y='SepalLengthCm', data=df, hue='Species')
plt.subplot(2, 2, 2)
sns.violinplot(x='Species', y='SepalWidthCm', data=df, hue='Species')
plt.subplot(2, 2, 3)
sns.violinplot(x='Species', y='PetalLengthCm', data=df, hue='Species')
plt.subplot(2, 2, 4)
sns.violinplot(x='Species', y='PetalWidthCm', data=df, hue='Species')
plt.tight_layout()
plt.show()
```

```python
# Histogram distribution of features
plt.figure(figsize=(15, 10))
plt.title('Feature Distribution using Histogram')
plt.subplot(2, 2, 1)
sns.histplot(df['SepalLengthCm'], kde=True, bins=30)
plt.subplot(2, 2, 2)
sns.histplot(df['SepalWidthCm'], kde=True, bins=30)
plt.subplot(2, 2, 3)
sns.histplot(df['PetalLengthCm'], kde=True, bins=30)
plt.subplot(2, 2, 4)
sns.histplot(df['PetalWidthCm'], kde=True, bins=30)
plt.tight_layout()
plt.show()
```

```python
# Correlation matrix
df_corr = df.drop('Species', axis=1)
plt.figure(figsize=(10, 5))
plt.title('Correlation Matrix')
sns.heatmap(df_corr.corr(), annot=True, cmap='coolwarm')
plt.show()
```

```python
# Scatter Plot of features
plt.figure(figsize=(15, 12))
plt.title('Scatter Plot of Features for Sepal Length and Sepal Width')

plt.subplot(1, 2, 1)
sns.scatterplot(x='SepalLengthCm', y='SepalWidthCm', hue='Species', data=df)

plt.subplot(1, 2, 2)
sns.scatterplot(x='PetalLengthCm', y='PetalWidthCm', hue='Species', data=df)

plt.tight_layout()
plt.show()
```

```python
# KDE Plot of features
plt.figure(figsize=(15, 12))
plt.title('KDE Plot of Features for Iris-versicolor')
versicolor = df[df['Species']=='Iris-versicolor']

plt.subplot(1, 2, 1)
sns.kdeplot(x=versicolor['SepalWidthCm'],y=versicolor['SepalLengthCm'],cmap='plasma',shade=True)

plt.subplot(1, 2, 2)
sns.kdeplot(x=versicolor['PetalWidthCm'],y=versicolor['PetalLengthCm'],cmap='plasma',shade=True)

plt.tight_layout()
plt.show()
```

```python
# KDE Plot of features
plt.figure(figsize=(15, 12))
plt.title('KDE Plot of Features for Sepal Length and Sepal Width for Iris-virginica')
virginica = df[df['Species']=='Iris-virginica']

plt.subplot(1, 2, 1)
sns.kdeplot(x=virginica['SepalWidthCm'],y=virginica['SepalLengthCm'],cmap='plasma',shade=True)

plt.subplot(1, 2, 2)
sns.kdeplot(x=virginica['PetalWidthCm'],y=virginica['PetalLengthCm'],cmap='plasma',shade=True)

plt.tight_layout()
plt.show()
```

```python
# KDE Plot of features
plt.figure(figsize=(15, 12))
plt.title('KDE Plot of Features for Sepal Length and Sepal Width for Iris-setosa')
setosa = df[df['Species']=='Iris-setosa']

plt.subplot(1, 2, 1)
sns.kdeplot(x=setosa['SepalWidthCm'],y=setosa['SepalLengthCm'],cmap='plasma',shade=True)

plt.subplot(1, 2, 2)
sns.kdeplot(x=setosa['PetalWidthCm'],y=setosa['PetalLengthCm'],cmap='plasma',shade=True)

plt.tight_layout()
plt.show()
```

```python
# Box Plot of features
plt.figure(figsize=(15, 10))
plt.title('Box Plot of Features')
plt.subplot(2, 2, 1)
sns.boxplot(x='Species', y='SepalLengthCm', data=df, hue='Species')
plt.subplot(2, 2, 2)
sns.boxplot(x='Species', y='SepalWidthCm', data=df, hue='Species')
plt.subplot(2, 2, 3)
sns.boxplot(x='Species', y='PetalLengthCm', data=df, hue='Species')
plt.subplot(2, 2, 4)
sns.boxplot(x='Species', y='PetalWidthCm', data=df, hue='Species')
plt.tight_layout()
plt.show()
```

```python
# Swarm Plot of features
plt.figure(figsize=(15, 10))
plt.title('Swarm Plot of Features')
plt.subplot(2, 2, 1)
sns.swarmplot(x='Species', y='SepalLengthCm', data=df, hue='Species')
plt.subplot(2, 2, 2)
sns.swarmplot(x='Species', y='SepalWidthCm', data=df, hue='Species')
plt.subplot(2, 2, 3)
sns.swarmplot(x='Species', y='PetalLengthCm', data=df, hue='Species')
plt.subplot(2, 2, 4)
sns.swarmplot(x='Species', y='PetalWidthCm', data=df, hue='Species')
plt.tight_layout()
plt.show()
```
